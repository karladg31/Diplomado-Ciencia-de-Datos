```python
medallas_por_disciplina = df.groupby('Discipline')['Medal_type'].count()

# Encontrar la disciplina con la mayor cantidad de medallas
disciplina_con_mas_medallas = medallas_por_disciplina.idxmax()
cantidad_medallas_maxima = medallas_por_disciplina.max()

print("La disciplina con mayor cantidad de medallas ganadas es '{}' con un total de {} medallas.".format(disciplina_con_mas_medallas, cantidad_medallas_maxima))
```

```python
medallas_por_disciplina = df.groupby('Discipline')['Medal_type'].count()

# Encontrar la disciplina con la menor cantidad de medallas
disciplina_con_menos_medallas = medallas_por_disciplina.idxmin()
cantidad_medallas_minima = medallas_por_disciplina.min()

print("La disciplina con menor cantidad de medallas ganadas es '{}' con un total de {} medallas.".format(disciplina_con_menos_medallas, cantidad_medallas_minima))
```

```python
medallas_por_pais = df.groupby('Committee')['Medal_type'].count()
top_tres_paises = medallas_por_pais.nlargest(3)
print("Los tres países con mayor cantidad de medallas ganadas son:")
for pais, cantidad_medallas in top_tres_paises.items():
    print("- {}: {} medallas".format(pais, cantidad_medallas))
tres_paises_menos_medallas = medallas_por_pais.nsmallest(3)
print("Los tres países con menor cantidad de medallas ganadas son:")
for pais, cantidad_medallas in tres_paises_menos_medallas.items():
    print("- {}: {} medallas".format(pais, cantidad_medallas))
```

```python
medallas_por_disciplina = df.groupby('Discipline')['Medal_type'].count().reset_index()
fig = go.Figure()
fig.add_trace(go.Bar(
    x=medallas_por_disciplina['Discipline'],
    y=medallas_por_disciplina['Medal_type'],
    marker_color='skyblue',
))
fig.update_layout(
    title='Cantidad de Medallas por Disciplina en los Juegos Olímpicos',
    xaxis=dict(title='Disciplina'),
    yaxis=dict(title='Cantidad de Medallas'),
)
fig.show()
```

```python
medallas_por_disciplina_y_tipo = df.groupby(['Discipline', 'Medal_type']).size().unstack(fill_value=0)
fig = go.Figure()
fig.add_trace(go.Bar(
    x=medallas_por_disciplina_y_tipo.index,
    y=medallas_por_disciplina_y_tipo['Bronze'],
    name='Bronze',
    marker_color='brown'
))

fig.add_trace(go.Bar(
    x=medallas_por_disciplina_y_tipo.index,
    y=medallas_por_disciplina_y_tipo['Silver'],
    name='Silver',
    marker_color='silver'
))

fig.add_trace(go.Bar(
    x=medallas_por_disciplina_y_tipo.index,
    y=medallas_por_disciplina_y_tipo['Gold'],
    name='Gold',
    marker_color='gold'
))
fig.update_layout(
    title='Cantidad de Medallas por Disciplina y Tipo en los Juegos Olímpicos',
    xaxis=dict(title='Disciplina'),
    yaxis=dict(title='Cantidad de Medallas'),
    barmode='stack',  # Para apilar las barras
)
fig.show()
```
