```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns #cosas estadísticas gg
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans #hoy
from sklearn.metrics import silhouette_score #hoy
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA #hoy
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.info()
```

```python
iris.describe()
```

```python
species_count = iris.species.value_counts().reset_index()
species_count.columns = ['Especie', 'Cuenta']
fig = px.bar(species_count, x = 'Especie', y = 'Cuenta', color = 'Especie', text = 'Cuenta', title = 'Distribución por especies', template = 'plotly_white')
fig.show()
```

```python
fig = go.Figure() #declara figura
for feature in iris.columns[:-1]: #para cada columna
  fig.add_trace(go.Box(y = iris[feature], x = iris.species, name = feature))
fig.update_layout(title = 'Gráfico de caja por característica', xaxis_title = 'Especie', yaxis_title = 'Valor', template = 'plotly_white')
fig.show()
```

```python
fig = px.scatter_matrix(iris, dimensions = iris.columns[:1], color = 'species', template = 'plotly_white')
fig.update_traces(diagonal_visible = False)
fig.update_layout(title = 'Pares de Características')
fig.show()
```

```python
X = iris.drop('species',axis = 1)
X
```

```python
wcss = [] #métrica para evaluar

for k in range (1,11):
  kmeans = KMeans(n_clusters = k, random_state = 42, n_init = 10)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

wcss
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = list(range(1, 11)), y = wcss, mode = 'lines+markers', name = 'WCSS'))
fig.update_layout(title = 'Método del Codo',
                  xaxis_title = 'Número de grupos (k)',
                  yaxis_title = 'WCSS',
                  xaxis = dict(tickmode = 'linear', tickvals = list(range(1, 11))),
                  yaxis = dict(rangemode = 'tozero'),
                  remplate = 'plotly_white')
fig.show()
```

```python
silhouette_scores = []

for k in range (2, 11):
  kmeans = KMeans(n_clusters = k, random_state = 42, n_init = 10)
  cluster_labels = kmeans.fit_predict(X)
  silhouette_scores.append(silhouette_score(X, cluster_labels))

silhouette_scores
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = list(range(2, 11)), y = silhouette_scores, mode = 'line+markers', name = 'Silhouette Score'))
fig.update_layout(title = 'Silhouette Score',
                  xaxis_title = 'Número de grupos (k)',
                  yaxis = dict (tickmode = 'linear', tickvals = list(range(2, 11))),
                  yaxis = dict(rangemode = 'tozero'),
                  template = 'plotly_white')
fig.show()
```
